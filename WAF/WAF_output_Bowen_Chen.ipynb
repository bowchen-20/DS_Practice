```python
#getting x and y
x = df.drop('outcome', axis = 1) 
y = df['outcome'] 
```

```python
from sklearn.feature_selection import f_regression
from sklearn.impute import SimpleImputer
#F-score
F_score, p_value = f_regression(x,y)
np.set_printoptions(suppress=True)
print(len(F_score))
print(F_score, df.columns)
```

```python
from sklearn.preprocessing import MinMaxScaler
#Now, lets scale all the value in x within 0 to 1...
scaler = MinMaxScaler()
scaler.fit(x)
x = scaler.transform(x)
x = pd.DataFrame(x)
```

```python
df.hist(bins=10, figsize=(20,20))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size = 0.2, random_state = 12)
X_train.shape, X_test.shape
```

```python
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,roc_curve,roc_auc_score
import matplotlib.pyplot as pltroc_auc_score

# Creating placer holders
ML_Model = []
acc_train = []
acc_test = []
f1_train = []
f1_test = []
recall_train = []
recall_test = []
pre_train = []
pre_test = []
auc_train = []
auc_test = []

#function to put results together
def storeResults(model,a,b,c,d,e,f,g,h,i,j):
  ML_Model.append(model)
  acc_train.append(round(a, 3))
  acc_test.append(round(b, 3))
  f1_train.append(round(c, 3))
  f1_test.append(round(d, 3))  
  recall_train.append(round(e, 3))
  recall_test.append(round(f, 3))  
  pre_train.append(round(g, 3))
  pre_test.append(round(h, 3))
  auc_train.append(round(i, 3))
  auc_test.append(round(j, 3))
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier

# instantiation 
tree = DecisionTreeClassifier()
# fit the model 
tree.fit(X_train, y_train)
# predicting the target value from the model for the samples
y_test_tree = tree.predict(X_test)
y_train_tree = tree.predict(X_train)
# finding accuracy and performance
acc_train_tree = accuracy_score(y_train,y_train_tree)
acc_test_tree = accuracy_score(y_test,y_test_tree)
f1_train_tree = f1_score(y_train,y_train_tree)
f1_test_tree = f1_score(y_test,y_test_tree)
recall_train_tree = recall_score(y_train,y_train_tree)
recall_test_tree = recall_score(y_test,y_test_tree)
pre_train_tree = precision_score(y_train,y_train_tree)
pre_test_tree = precision_score(y_test,y_test_tree)
auc_train_tree =roc_auc_score(y_train,y_train_tree)
auc_test_tree = roc_auc_score(y_test,y_test_tree)
print("Decision Tree: Accuracy on training Data: {:.3f}".format(acc_train_tree))
print("Decision Tree: Accuracy on test Data: {:.3f}".format(acc_test_tree))
print("Decision Tree: f1 on training Data: {:.3f}".format(f1_train_tree))
print("Decision Tree: f1 on test Data: {:.3f}".format(f1_test_tree))
print("Decision Tree: recall on training Data: {:.3f}".format(recall_train_tree))
print("Decision Tree: recall on test Data: {:.3f}".format(recall_test_tree))
print("Decision Tree: precision on training Data: {:.3f}".format(pre_train_tree))
print("Decision Tree: precision on test Data: {:.3f}".format(pre_test_tree))
print("Decision Tree: AUC on training Data: {:.3f}".format(auc_train_tree))
print("Decision Tree: AUC on test Data: {:.3f}".format(auc_test_tree))
```

```python
storeResults('Decision Tree',acc_train_tree,acc_test_tree,f1_train_tree,f1_test_tree,recall_train_tree,
             recall_test_tree,pre_train_tree,pre_test_tree,auc_train_tree,auc_test_tree)
```

```python
# Random Forest model
from sklearn.ensemble import RandomForestClassifier

# instantiation
forest = RandomForestClassifier()

# fit the model 
forest.fit(X_train, y_train)
# predicting the target value from the model for the samples
y_test_forest = forest.predict(X_test)
y_train_forest = forest.predict(X_train)
# finding accuracy and performance
acc_train_forest = accuracy_score(y_train,y_train_forest)
acc_test_forest = accuracy_score(y_test,y_test_forest)
f1_train_forest = f1_score(y_train,y_train_forest)
f1_test_forest = f1_score(y_test,y_test_forest)
recall_train_forest = recall_score(y_train,y_train_forest)
recall_test_forest = recall_score(y_test,y_test_forest)
pre_train_forest = precision_score(y_train,y_train_forest)
pre_test_forest = precision_score(y_test,y_test_forest)
auc_train_forest = roc_auc_score(y_train,y_train_forest)
auc_test_forest = roc_auc_score(y_test,y_test_forest)
print("Random forest: Accuracy on training Data: {:.3f}".format(acc_train_forest))
print("Random forest: Accuracy on test Data: {:.3f}".format(acc_test_forest))
print("Random forest: f1 on training Data: {:.3f}".format(f1_train_forest))
print("Random forest: f1 on test Data: {:.3f}".format(f1_test_forest))
print("Random forest: recall on training Data: {:.3f}".format(recall_train_forest))
print("Random forest: recall on test Data: {:.3f}".format(recall_test_forest))
print("Random forest: precision on training Data: {:.3f}".format(pre_train_forest))
print("Random forest: precision on test Data: {:.3f}".format(pre_test_forest))
print("Random forest: AUC on training Data: {:.3f}".format(pre_train_forest))
print("Random forest: AUC on test Data: {:.3f}".format(pre_test_forest))
```

```python
storeResults('Random Forest', acc_train_forest, acc_test_forest,f1_train_forest,f1_test_forest,recall_train_forest,
             recall_test_forest,pre_train_forest,pre_test_forest,auc_train_forest,auc_test_forest)
```
